```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
import plotly.graph_objects as go
from plotly.subplots import make_subplots
df=pd.read_csv('consolidated_all_scenarios.csv')
# Ensure 'Count' is numeric
df['Count'] = pd.to_numeric(df['Count'])
```

```python
df=df.dropna()
df
```

```python
df['Input'] = df['Input'].str.replace('water_td_irr_', '')
df['Input'] = df['Input'].str.replace('_W', '')
```

```python

df['Combination'] = df['Combination'].replace({'FiberCrop': 'Cotton', 'SugarCrop': 'Sugarcane', 'MiscAll': 'Miscellaneous'}, regex=True)
df
```

```python
# Extract crop names from combinations
df['Crop1'] = df['Combination'].apply(lambda x: x.split(' vs ')[0].split('_')[0])
df['Crop2'] = df['Combination'].apply(lambda x: x.split(' vs ')[1].split('_')[0])


# Setting global plot parameters
plt.rcParams.update({
    'axes.titlesize': 16,     # Title font size
    'axes.labelsize': 14,     # Axis label font size
    'xtick.labelsize': 14,    # X-axis tick label font size
    'ytick.labelsize': 14,    # Y-axis tick label font size
    'legend.fontsize': 14,    # Legend font size
    'figure.titlesize': 18    # Figure title font size
})


# Create a new column with sorted crop pairs
df['CropPair'] = df.apply(lambda row: '_'.join(sorted([row['Crop1'], row['Crop2']])), axis=1)

# Create a pivot table with the new CropPair column
pivot_data_crops = df.pivot_table(values='Count', index='Input', columns='CropPair', aggfunc='sum', fill_value=0)

# Calculate total counts for rows and columns separately
row_totals = pivot_data_crops.sum(axis=1)
col_totals = pivot_data_crops.sum(axis=0)

# Create the heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(pivot_data_crops, annot=True, fmt='d', cmap='YlOrRd')
plt.title('19_BOI_Outlying_Scenarios')
plt.xlabel('Crop Pair')
plt.ylabel('Region')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('19_BOI_Outlying_Scenarios.png', dpi=600, bbox_inches='tight')
plt.show()


print("Row Totals:")
print(row_totals)

print("\nColumn Totals:")
print(col_totals)
```

```python
##code from abby
import os
import pandas as pd

def expanding_scenario(df):
    df[['var1','var2','var3','gw','res','esm','tax']] = df['Scenario'].str.split('_', n=6, expand=True)
    df.loc[df.gw=='gwlo','gw']='1'
    df.loc[df.gw=='gwmed','gw']='2'
    df.loc[df.gw=='gwhi','gw']='3'
    df.loc[df.esm=='gfdl','esm']='1'
    df.loc[df.esm=='hadgem','esm']='2'
    df.loc[df.esm=='ipsl','esm']='3'
    df.loc[df.esm=='miroc','esm']='4'
    df.loc[df.esm=='noresm','esm']='5'
    df.loc[df.res=='rs','res']='1'
    df.loc[df.res=='exp','res']='2'
    df.loc[df.tax=='NDC_Tax_ffict','tax']='1'
    df.loc[df.tax=='NDC_Tax_uct','tax']='2'
    df['ssp'], df['ag'], df['soc'] = '', '', ''
    df.loc[df.var2.str[0]=='a','ssp']='2'
    df.loc[df.var2.str[0]=='s','ssp']='1'
    df.loc[df.var2.str[0]=='a','ag']=df['var2'].str[-1]
    df.loc[df.var2.str[0]=='s','ag']=df['var3'].str[-1]
    df.loc[df.var3.str[0]=='s','soc']=df['var3'].str[-1]
    df.loc[df.var3.str[0]=='a','soc']=df['var2'].str[-1]
    df.drop(['var1', 'var2', 'var3'], axis=1, inplace=True)
    df[['ssp', 'ag', 'soc', 'esm', 'tax', 'res', 'gw']] = df[['ssp', 'ag', 'soc', 'esm', 'tax', 'res', 'gw']].astype(int)
    return df


```

```python
expanding_scenario(df)
```

```python
### memory intensive and not required (optional to explore will generate approx 1500 heatmaps)
scenarios = df['Scenario'].unique()

# Create the target directory if it does not exist
output_directory = 'unique_scenario_heatmaps_count'
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Create a heatmap for each scenario
for scenario in scenarios:
    scenario_data = df[df['Scenario'] == scenario]
    pivot_data = scenario_data.pivot_table(values='Count', index='Input', columns='Combination', aggfunc='sum', fill_value=0)
    
    plt.figure(figsize=(20, 12))
    sns.heatmap(pivot_data, annot=True, fmt='d', cmap='YlOrRd')
    plt.title(f'Heatmap of Combinations Across Regions for Scenario: {scenario}')
    plt.xlabel('Combination')
    plt.ylabel('Region')
    plt.xticks(rotation=90)
    plt.tight_layout()
    
    # Save the figure to the specified directory
    output_file_path = os.path.join(output_directory, f'heatmap_combinations_regions_{scenario}.png')
    plt.savefig(output_file_path, dpi=300, bbox_inches='tight')
    plt.close()
    
    print(f'Saved heatmap for scenario {scenario} to {output_file_path}')

print("All heatmaps have been generated and saved.")
```
