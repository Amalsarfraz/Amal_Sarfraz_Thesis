```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import pandas as pd 
import seaborn as sns 
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
matplotlib.rcParams['font.sans-serif'] = "Calibri"
matplotlib.rcParams['font.family'] = "sans-serif"
sns.set_context("poster", rc={"axes.titlesize":18, "axes.labelsize":18})
colors = sns.color_palette("colorblind")  # This will fetch a set of colors designed for colorblindness
```

```python
#############naming renaming scenario

#####################UPDATEDDDDD########################from Abby
def expanding_scenario(df):
# turn scenario into columns for different important dimensions

    df[['var1','var2','var3','gw','res','esm','tax']] = df['scenario'].str.split('_',n=6,expand=True)

    # #replacing scenario names with numerical conventions

    #groundwater
    df.loc[df.gw=='gwlo','gw']='1'
    df.loc[df.gw=='gwmed','gw']='2'
    df.loc[df.gw=='gwhi','gw']='3'

    #earth systems model
    df.loc[df.esm=='gfdl','esm']='1'
    df.loc[df.esm=='hadgem','esm']='2'
    df.loc[df.esm=='ipsl','esm']='3'
    df.loc[df.esm=='miroc','esm']='4'
    df.loc[df.esm=='noresm','esm']='5'

    #reservoir
    df.loc[df.res=='rs','res']='1'
    df.loc[df.res=='exp','res']='2'

    #tax
    df.loc[df.tax=='NDC_Tax_ffict','tax']='1'
    df.loc[df.tax=='NDC_Tax_uct','tax']='2'

    #now dealing with the first few variables...

    #add empty columns for ssp, ag, and soc
    df['ssp']=''
    df['ag']=''
    df['soc']=''

    #give var1 (ssp) value of 1 if var2 = socio and var3 = ag and value of 2 if var2 = ag and var3 = socio
    df.loc[df.var2.str[0]=='a','ssp']='2'
    df.loc[df.var2.str[0]=='s','ssp']='1'

    #assign numberic value to ag for which ag ssp it is - 
    df.loc[df.var2.str[0]=='a','ag']=df['var2'].str[-1]
    df.loc[df.var2.str[0]=='s','ag']=df['var3'].str[-1]

    #asign numberic value to soc for which soc ssp it is - 
    df.loc[df.var3.str[0]=='s','soc']=df['var3'].str[-1]
    df.loc[df.var3.str[0]=='a','soc']=df['var2'].str[-1]

    #get rid of var1, var2, var3 columns
    df = df.drop(['var1','var2','var3'],axis=1)

    #make the soc, ag, and ssp columns to be int not str
    df['ssp'] = df['ssp'].astype(int)
    df['soc'] = df['soc'].astype(int)
    df['ag'] = df['ag'].astype(int)
    df['esm'] = df['esm'].astype(int)
    df['tax'] = df['tax'].astype(int)
    df['res'] = df['res'].astype(int)
    df['gw'] = df['gw'].astype(int)
    
    return df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df_60_scenarios=pd.read_pickle('unique_sectors/df_scenario_60.pkl')
df_60_scenarios
scenario_60_names = df_60_scenarios['scenario'].unique()

dataframes = ["Misc_irri", "FiberCrop_irri", "Wheat_irri", "Rice_irri", "SugarCrop_irri"]
crop_dfs = {name: pd.read_pickle(f'unique_sectors/{name}.pkl') for name in dataframes}

df_misc, df_cotton, df_wheat, df_rice, df_sugar = crop_dfs.values()

def plot_and_save_all_and_scenario_60_stats(crop_dfs, crop_names, scenario_60_names):
    # Set the general font size
    plt.rcParams.update({'font.size': 30})
    
    # Calculate the overall maximum value across all crops for consistent y-axis limits
    max_irri_ww = max(
        crop_df['irri_ww'].max() for crop_df in crop_dfs
    )
    
    # Setting up the figure for 2x3 subplots (2 rows, 3 columns)
    fig, axs = plt.subplots(2, 3, figsize=(30, 15))  # Adjust the figure size as needed
    
    crop_types = ['Summer Crop', 'Summer Crop', 'Winter Crop', 'Annual Crop', 'Annual Crop']  # Matching each crop with its type
    
    colors = sns.color_palette("colorblind")  # This will fetch a set of colors designed for colorblindness
    
    for i, (crop_df, crop_name, crop_type) in enumerate(zip(crop_dfs, crop_names, crop_types)):
        # Determine the row and column indices for the subplot
        row = i // 3
        col = i % 3

        # All scenarios stats
        all_stats = crop_df.groupby('year').agg({'irri_ww': ['max', 'min', 'mean']})
        
        # Scenario 60 stats
        scenario_60_df = crop_df[crop_df['scenario'].isin(scenario_60_names)]
        scenario_60_stats = scenario_60_df.groupby('year').agg({'irri_ww': ['max', 'min', 'mean']})

        # Plotting on the ith subplot using the colorblind-friendly palette
        ax = axs[row, col]

        ax.axvspan(2015, 2050, color='#9C9B7A', alpha=0.3)
        ax.axvspan(2050, 2100, color='yellow', alpha=0.3)
        
        ax.plot(all_stats.index, all_stats['irri_ww']['max'], label='3000 Scenarios Max', color=colors[0], linewidth=3)
        ax.plot(all_stats.index, all_stats['irri_ww']['min'], label='3000 Scenarios Min', color=colors[1], linewidth=3)
        ax.plot(all_stats.index, all_stats['irri_ww']['mean'], label='3000 Scenarios Mean', color=colors[2], linewidth=3)
        ax.plot(scenario_60_stats.index, scenario_60_stats['irri_ww']['max'], label='OS5 60 scenarios Max', linestyle=':', color=colors[0], linewidth=4)
        ax.plot(scenario_60_stats.index, scenario_60_stats['irri_ww']['min'], label='OS5 60 scenarios Min', linestyle=':', color=colors[1], linewidth=4)
        ax.plot(scenario_60_stats.index, scenario_60_stats['irri_ww']['mean'], label='O5 60 scenarios Mean', linestyle=':', color=colors[2], linewidth=4)

        ax.set_title(f'{crop_name} ({crop_type})', fontsize=30, weight='bold')
        ax.set_xlabel('Year', fontsize=30)
        ax.set_ylabel('Irrigation Withdrawal (km$^{\mathbf{3}}$)', fontsize=30, weight='bold')
        ax.tick_params(labelbottom=True, labelsize=25)
        ax.set_ylim(0, max_irri_ww * 1.1)
    
    # Remove empty subplot (last one in the second row)
    fig.delaxes(axs[1, 2])
    
    # Adjust layout and add spacing between rows
    plt.subplots_adjust(hspace=0.4, wspace=0.3)
    
    # Create a combined legend for both the shaded regions and the lines
    custom_legend = [
        plt.Line2D([0], [0], color='#9C9B7A', alpha=0.3, linewidth=10, label='2015-2050'),
        plt.Line2D([0], [0], color='yellow', alpha=0.3, linewidth=10, label='2050-2100'),
        plt.Line2D([0], [0], color=colors[0], linewidth=3, label='3000 Scenarios Max'),
        plt.Line2D([0], [0], linestyle=':', color=colors[0], linewidth=4, label='OS5 60 scenarios Max'),
        plt.Line2D([0], [0], color=colors[1], linewidth=3, label='3000 Scenarios Min'),
        plt.Line2D([0], [0], linestyle=':', color=colors[1], linewidth=4, label='OS5 60 scenarios Min'),
        plt.Line2D([0], [0], color=colors[2], linewidth=3, label='3000 Scenarios Mean'),
        plt.Line2D([0], [0], linestyle=':', color=colors[2], linewidth=4, label='OS5 60 scenarios Mean')
    ]
    fig.legend(handles=custom_legend, loc='upper center', bbox_to_anchor=(0.78, 0.45), fontsize=30, ncol=1)
    
    #plt.savefig("Irrigation_Stats_60_3000.png", bbox_inches='tight', dpi=900)
    plt.show()

crop_dfs = [df_cotton, df_rice, df_wheat, df_sugar, df_misc]
crop_names = ['Cotton', 'Rice', 'Wheat', 'Sugarcane', 'Miscellaneous']
plot_and_save_all_and_scenario_60_stats(crop_dfs, crop_names, scenario_60_names)
```
