```python
### getting to unique patterns visualising frequency etc 
import csv
from collections import defaultdict
import matplotlib.pyplot as plt

def parse_values(value_str):
    return tuple(sorted(map(int, value_str.replace(" ", "").split(","))))

def analyze_patterns(file_path):
    patterns = defaultdict(list)
    
    with open(file_path, 'r') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            pattern = (
                parse_values(row['tax']),
                parse_values(row['ssp']),
                parse_values(row['ag']),
                parse_values(row['soc'])
            )
            patterns[pattern].append(row['FileName'])
    
    return patterns

def get_variable_frequency(patterns):
    var_freq = {
        'tax': defaultdict(int),
        'ssp': defaultdict(int),
        'ag': defaultdict(int),
        'soc': defaultdict(int)
    }
    
    for pattern, files in patterns.items():
        count = len(files)
        var_freq['tax'][pattern[0]] += count
        var_freq['ssp'][pattern[1]] += count
        var_freq['ag'][pattern[2]] += count
        var_freq['soc'][pattern[3]] += count
    
    return var_freq

def plot_variable_frequency(var_freq):
    fig, axs = plt.subplots(2, 2, figsize=(15, 15))
    fig.suptitle('Variable Frequency Analysis', fontsize=16)

    for i, (var, freq) in enumerate(var_freq.items()):
        ax = axs[i // 2, i % 2]
        labels, values = zip(*sorted(freq.items(), key=lambda x: x[1], reverse=True))
        ax.bar(range(len(labels)), values)
        ax.set_xticks(range(len(labels)))
        ax.set_xticklabels([str(l) for l in labels], rotation=45)
        ax.set_title(f'{var.upper()} Frequency')
        ax.set_ylabel('Count')

    plt.tight_layout()
    #plt.savefig('variable_frequency.png')
    plt.show()

def plot_pattern_frequency(patterns):
    pattern_freq = sorted([(p, len(f)) for p, f in patterns.items()], key=lambda x: x[1], reverse=True)
    top_patterns = pattern_freq[:]  # Top 10 patterns

    labels = [f"P{i+1}" for i in range(len(top_patterns))]
    values = [count for _, count in top_patterns]

    plt.figure(figsize=(12, 6))
    plt.bar(labels, values)
    plt.title('Top 10 Pattern Frequencies')
    plt.xlabel('Pattern')
    plt.ylabel('Count')
    #plt.savefig('pattern_frequency.png')
    plt.show()

    print("All Patterns:")
    for i, (pattern, count) in enumerate(top_patterns, 1):
        print(f"P{i}: TAX{pattern[0]} SSP{pattern[1]} AG{pattern[2]} SOC{pattern[3]} - Count: {count}")

# Usage
file_path = 'consolidated_sets_scenarios_92.csv'
patterns = analyze_patterns(file_path)
var_freq = get_variable_frequency(patterns)

plot_variable_frequency(var_freq)
plot_pattern_frequency(patterns)

print("\nPlots have been saved as 'variable_frequency.png' and 'pattern_frequency.png'")
```

```python
### coping pattern file to respective folder creating pattern folders 
import csv
import shutil
import os
from collections import defaultdict

def parse_values(value_str):
    return tuple(sorted(map(int, value_str.replace(" ", "").split(","))))

def analyze_patterns(file_path, source_directory):
    patterns = defaultdict(list)
    with open(file_path, 'r') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            pattern = (
                parse_values(row['tax']),
                parse_values(row['ssp']),
                parse_values(row['ag']),
                parse_values(row['soc'])
            )
            # Construct full file path for each .csv and .txt file
            for ext in ['.csv', '.txt']:
                file_path = os.path.join(source_directory, f"{row['FileName']}{ext}")
                if os.path.exists(file_path):  # Check if the file actually exists
                    patterns[pattern].append((file_path, int(row['NumScenarios'])))
    return patterns

def ensure_dir(file_path):
    if not os.path.exists(file_path):
        os.makedirs(file_path)

def copy_files_to_pattern_folder(patterns):
    base_path = 'pattern_folders'
    ensure_dir(base_path)  # Ensure the base directory exists
    
    # Create and populate pattern folders
    for i, (pattern, files) in enumerate(sorted(patterns.items(), key=lambda x: len(x[1]), reverse=True), start=1):
        pattern_name = f"Pattern {i}"
        pattern_dir = os.path.join(base_path, pattern_name)
        ensure_dir(pattern_dir)
        
        for file_path, _ in files:
            file_name = os.path.basename(file_path)
            dest = os.path.join(pattern_dir, file_name)
            shutil.copy2(file_path, dest)  # Copy file to the pattern directory
            print(f"Copied {file_path} to {dest}")  # Debug: Output the operation

# Usage
source_directory = 'C:/Users/HP/Desktop/OSTI_Application_DD_Water_Basins/7. BOI_OSTI_29_June/OS_overlap_analysis/93_outlier_sets'  # Define your source directory here
file_path = 'consolidated_sets_scenarios_92.csv'
patterns = analyze_patterns(file_path, source_directory)

# Ensure there are files and patterns found
if patterns:
    copy_files_to_pattern_folder(patterns)
    print("Files have been organized into pattern folders.")
else:
    print("No patterns found or no files matched the criteria.")
```

```python
## saving the unique patterns in a csv file for further screening
import csv
from collections import defaultdict

def parse_values(value_str):
    return tuple(sorted(map(int, value_str.replace(" ", "").split(","))))

def analyze_patterns(file_path):
    patterns = defaultdict(list)
    
    with open(file_path, 'r') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            pattern = (
                parse_values(row['tax']),
                parse_values(row['ssp']),
                parse_values(row['ag']),
                parse_values(row['soc'])
            )
            patterns[pattern].append((row['FileName'], int(row['NumScenarios'])))
    
    return patterns

def print_all_patterns(patterns):
    print("All Patterns:")
    for i, (pattern, files) in enumerate(sorted(patterns.items(), key=lambda x: len(x[1]), reverse=True), 1):
        print(f"\nPattern {i}:")
        print(f"  TAX: {pattern[0]}")
        print(f"  SSP: {pattern[1]}")
        print(f"  AG: {pattern[2]}")
        print(f"  SOC: {pattern[3]}")
        print(f"  Occurrences: {len(files)}")
        total_scenarios = sum(file[1] for file in files)
        print(f"  Total Scenarios: {total_scenarios}")
        print("  Files:")
        for file, num_scenarios in files:
            print(f"    - {file} (Scenarios: {num_scenarios})")

def get_variable_frequency(patterns):
    var_freq = {
        'tax': defaultdict(int),
        'ssp': defaultdict(int),
        'ag': defaultdict(int),
        'soc': defaultdict(int)
    }
    
    for pattern, files in patterns.items():
        count = len(files)
        var_freq['tax'][pattern[0]] += count
        var_freq['ssp'][pattern[1]] += count
        var_freq['ag'][pattern[2]] += count
        var_freq['soc'][pattern[3]] += count
    
    return var_freq

def print_variable_frequency(var_freq):
    for var, freq in var_freq.items():
        print(f"\n{var.upper()} Frequency:")
        for value, count in sorted(freq.items(), key=lambda x: x[1], reverse=True):
            print(f"  {value}: {count}")

def save_patterns_to_csv(patterns, output_file):
    with open(output_file, 'w', newline='') as csvfile:
        fieldnames = ['Pattern_Number', 'TAX', 'SSP', 'AG', 'SOC', 'Occurrences', 'Total_Scenarios', 'Files']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        
        writer.writeheader()
        for i, (pattern, files) in enumerate(sorted(patterns.items(), key=lambda x: len(x[1]), reverse=True), 1):
            total_scenarios = sum(file[1] for file in files)
            writer.writerow({
                'Pattern_Number': i,
                'TAX': pattern[0],
                'SSP': pattern[1],
                'AG': pattern[2],
                'SOC': pattern[3],
                'Occurrences': len(files),
                'Total_Scenarios': total_scenarios,
                'Files': "; ".join([f"{file} (Scenarios: {num_scenarios})" for file, num_scenarios in files])
            })

# Usage
file_path = 'consolidated_sets_scenarios_92.csv'
output_file = 'unique_patterns_from_92_OS.csv'

patterns = analyze_patterns(file_path)
print_all_patterns(patterns)

var_freq = get_variable_frequency(patterns)
print_variable_frequency(var_freq)

save_patterns_to_csv(patterns, output_file)

print(f"\nTotal unique patterns: {len(patterns)}")
```

```python
import pandas as pd
data=pd.read_csv('unique_patterns_from_92_OS.csv')
# Adjusting the parsing function to handle empty strings more gracefully
def parse_tuple_string_v2(tuple_str):
    items = tuple_str.strip("()").split(',')
    return tuple(int(item) for item in items if item)

# Re-apply the parsing function to the columns
data['TAX'] = data['TAX'].apply(parse_tuple_string_v2)
data['SSP'] = data['SSP'].apply(parse_tuple_string_v2)
data['AG'] = data['AG'].apply(parse_tuple_string_v2)
data['SOC'] = data['SOC'].apply(parse_tuple_string_v2)

# Now, identify which patterns (with fewer scenarios) are a subset of others (with more scenarios)
subset_patterns = []

# Iterate over each pattern and check if it is a subset of any other pattern
for i, pattern1 in data.iterrows():
    for j, pattern2 in data.iterrows():
        if i != j:
            if (set(pattern1['TAX']).issubset(set(pattern2['TAX'])) and
                set(pattern1['SSP']).issubset(set(pattern2['SSP'])) and
                set(pattern1['AG']).issubset(set(pattern2['AG'])) and
                set(pattern1['SOC']).issubset(set(pattern2['SOC']))):
                subset_patterns.append((pattern1['Pattern_Number'], pattern2['Pattern_Number']))

subset_patterns_df = pd.DataFrame(subset_patterns, columns=['Subset Pattern', 'Superset Pattern'])

#import ace_tools as tools; tools.display_dataframe_to_user(name="Subset and Superset Patterns", dataframe=subset_patterns_df)

subset_patterns_df
```

```python
import csv
from collections import defaultdict, Counter

def parse_values(value_str):
    return tuple(sorted(map(int, value_str.replace(" ", "").split(","))))

def analyze_patterns(file_path):
    patterns = []
    file_scenarios = {}
    
    with open(file_path, 'r') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            pattern = (
                parse_values(row['tax']),
                parse_values(row['ssp']),
                parse_values(row['ag']),
                parse_values(row['soc'])
            )
            patterns.append(pattern)
            file_scenarios[row['FileName']] = int(row['NumScenarios'])
    
    return patterns, file_scenarios

def count_combinations(patterns):
    return Counter(patterns)

def print_analysis(patterns, file_scenarios, combinations):
    print("Analysis based on filename:")
    print(f"Total unique filenames: {len(file_scenarios)}")
    print(f"Total scenarios across all files: {sum(file_scenarios.values())}")
    
    print("\nNumber of scenarios per file:")
    for filename, num_scenarios in file_scenarios.items():
        print(f"  {filename}: {num_scenarios}")
    
    print("\nUnique combinations and their frequencies:")
    for combo, freq in combinations.items():
        print(f"  Combination: TAX{combo[0]} SSP{combo[1]} AG{combo[2]} SOC{combo[3]} - Frequency: {freq}")
    
    print(f"\nTotal unique combinations: {len(combinations)}")
    
    print("\nVariable analysis:")
    variables = ['tax', 'ssp', 'ag', 'soc']
    for i, var in enumerate(variables):
        values = Counter(pattern[i] for pattern in patterns)
        print(f"\n  {var.upper()} values and their frequencies:")
        for value, freq in values.items():
            print(f"    {value}: {freq}")

# Usage
file_path = 'consolidated_sets_scenarios_92.csv'
patterns, file_scenarios = analyze_patterns(file_path)
combinations = count_combinations(patterns)
print_analysis(patterns, file_scenarios, combinations)
```

```python
import csv
from collections import defaultdict
import os

def analyze_patterns(file_path):
    patterns = defaultdict(list)
    
    with open(file_path, 'r') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            pattern = (
                tuple(sorted(map(int, row['tax'].replace(" ", "").split(",")))),
                tuple(sorted(map(int, row['ssp'].replace(" ", "").split(",")))),
                tuple(sorted(map(int, row['ag'].replace(" ", "").split(",")))),
                tuple(sorted(map(int, row['soc'].replace(" ", "").split(","))))
            )
            patterns[pattern].append(row['FileName'])
    
    return patterns

def get_scenarios_from_file(base_folder, filename):
    scenarios = set()
    # Remove '_scenarios' from the filename if present
    filename = filename.replace('_scenarios', '')
    full_path = os.path.join(base_folder, '93_outlier_sets', '93_csv_only', filename + '.csv')
    try:
        with open(full_path, 'r') as f:
            reader = csv.DictReader(f)
            for row in reader:
                scenarios.add(row['scenario'])  # 
    except FileNotFoundError:
        print(f"Warning: File {full_path} not found.")
    return scenarios

def analyze_overlapping_scenarios(patterns, base_folder):
    overlaps = defaultdict(lambda: defaultdict(set))
    
    combinations = [combo for combo, files in patterns.items() if len(files) > 1]
    
    for i, combo1 in enumerate(combinations):
        files1 = patterns[combo1]
        scenarios1 = set()
        for file in files1:
            scenarios1.update(get_scenarios_from_file(base_folder, file))
        
        for combo2 in combinations[i+1:]:
            files2 = patterns[combo2]
            scenarios2 = set()
            for file in files2:
                scenarios2.update(get_scenarios_from_file(base_folder, file))
            
            overlap = scenarios1.intersection(scenarios2)
            if overlap:
                overlaps[combo1][combo2] = overlap
                overlaps[combo2][combo1] = overlap
    
    return overlaps

def print_analysis(patterns, overlaps):
    print("Combinations appearing more than once:")
    for combo, files in patterns.items():
        if len(files) > 1:
            print(f"\nCombination: TAX{combo[0]} SSP{combo[1]} AG{combo[2]} SOC{combo[3]}")
            print(f"Frequency: {len(files)}")
            print("Files:")
            for file in files:
                print(f"  - {file}")
    
    print("\nOverlapping scenarios between combinations:")
    for combo1, others in overlaps.items():
        print(f"\nOverlaps for TAX{combo1[0]} SSP{combo1[1]} AG{combo1[2]} SOC{combo1[3]}:")
        for combo2, overlap in others.items():
            if overlap:
                print(f"  With TAX{combo2[0]} SSP{combo2[1]} AG{combo2[2]} SOC{combo2[3]}:")
                print(f"    {len(overlap)} overlapping scenarios: {', '.join(sorted(overlap))}")

# Usage
file_path = 'consolidated_sets_scenarios_92.csv'
base_folder = 'C:/Users/HP/Desktop/OSTI_Application_DD_Water_Basins/7. BOI_OSTI_29_June/OS_overlap_analysis/'
patterns = analyze_patterns(file_path)
overlaps = analyze_overlapping_scenarios(patterns, base_folder)
print_analysis(patterns, overlaps)
```
